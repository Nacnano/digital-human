```python
import pandas as pd

xls = pd.ExcelFile('data/Video List.xlsx')
human = pd.read_excel(xls, 'AVG')
gemini = pd.read_excel(xls, 'GENIE')
```

```python
columns = ["Speed", "Naturalness", "Continuity", "Listening Effort", "Eye contact", "Posture", "Hand gesture"]

# Calculate Kendall Tau correlation for each column
for col in columns:
    tau, p_value = pd.Series(human[col]).corr(pd.Series(gemini[col]), method='kendall'), None
    print(f"{col}: Kendall Tau = {tau:.4f}")
    
```

```python
# Calculate Spearman correlation for each column

for col in columns:
    spearman, p_value = pd.Series(human[col]).corr(pd.Series(gemini[col]), method='spearman'), None
    print(f"{col}: Spearman = {spearman:.4f}")
```
